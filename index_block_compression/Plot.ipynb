```python
def plot_cr_ra(Baseline_x, Baseline_y, Leco_x, Leco_y, title,scale):
    baseline, = plt.plot(Baseline_x, Baseline_y, label = 'baseline')
    plt.scatter(Baseline_x, Baseline_y)
    leco, = plt.plot(Leco_x, Leco_y, label = 'leco')
    plt.legend(handles=[baseline, leco])
    plt.scatter(Leco_x, Leco_y)
    plt.xlim(0, scale)
    plt.ylim(0,100)
    plt.xlabel('random access')
    plt.ylabel('compression rate')
    plt.title(title)
    plt.show()
```
